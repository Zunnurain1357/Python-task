### 6
### peak hours of sales

```python
df['sales'] = df['quantity'] * df['price']
df['sales']
```

```python
df['hours'] = df['time'].dt.hour
```

```python
df['hours'] 
```

```python
hourly_sales = df.groupby('hours')['sales'].sum().reset_index()
hourly_sales
```

```python
peak_hour = hourly_sales.loc[hourly_sales['sales'].idxmax()]

peak_hour
print(f"peak hour: {peak_hour['hours']}:00 with ${peak_hour['sales']:.2f} in sales")
```

```python
plt.figure(figsize=(10,6))
plt.bar(hourly_sales['hours'],hourly_sales['sales'])
plt.xlabel('hours')
plt.ylabel('sales')
plt.title('hourly_sales')
plt.show()
```

### 7
## Total sales made on each day of the week 

```python
df['DayOfWeek'] = df['date'].dt.day_name()
df['DayOfWeek']
```

```python
daily_sales = df.groupby('DayOfWeek')['sales'].sum().reset_index()

daily_sales
```

```python
plt.figure(figsize=(10,6))
plt.bar(daily_sales['DayOfWeek'],daily_sales['sales'])
plt.xlabel('Day Of Week')
plt.ylabel('sales')
plt.title('Tota sales of week')
plt.show()
```
